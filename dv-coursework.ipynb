```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Update file path based on Kaggle's environment)
file_path = "/kaggle/working/cleaned_dataset.csv"  # Update this with your file path
df = pd.read_csv(file_path)

# Drop rows where Latitude or Longitude are missing
df_cleaned = df.dropna(subset=["Latitude", "Longitude"])

# Set seaborn style
sns.set_style("whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# 1. Histogram of Rent Prices
sns.histplot(df_cleaned["Rent"], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Rent Prices")
axes[0, 0].set_xlabel("Rent (AED)")
axes[0, 0].set_ylabel("Count")

# 2. Rent Category Count
sns.countplot(x=df_cleaned["Rent_category"], palette="viridis", ax=axes[0, 1])
axes[0, 1].set_title("Number of Listings by Rent Category")
axes[0, 1].set_xlabel("Rent Category")
axes[0, 1].set_ylabel("Count")

# 3. Beds and Baths Distribution
sns.countplot(x=df_cleaned["Beds"], ax=axes[1, 0], palette="coolwarm")
axes[1, 0].set_title("Distribution of Bedroom Counts")

sns.countplot(x=df_cleaned["Baths"], ax=axes[1, 1], palette="coolwarm")
axes[1, 1].set_title("Distribution of Bathroom Counts")

# 4. Rent vs Area Scatter Plot
sns.scatterplot(x=df_cleaned["Area_in_sqft"], y=df_cleaned["Rent"], ax=axes[2, 0])
axes[2, 0].set_title("Rent vs Area in Sqft")
axes[2, 0].set_xlabel("Area (Sqft)")
axes[2, 0].set_ylabel("Rent (AED)")

# 5. Rent Distribution by Furnishing Type
sns.boxplot(x="Furnishing", y="Rent", data=df_cleaned, ax=axes[2, 1])
axes[2, 1].set_title("Rent Distribution by Furnishing Type")
axes[2, 1].set_xlabel("Furnishing Type")
axes[2, 1].set_ylabel("Rent (AED)")

# Adjust layout
plt.tight_layout()
plt.show()

# Display correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
# Select only numeric columns for correlation matrix
numeric_cols = df_cleaned.select_dtypes(include=['number'])

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
